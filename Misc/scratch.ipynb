```python
def partition(arr, lb, ub):
    start = lb
    end = ub
    p = arr[lb]  # Corrected from 'a' to 'arr'
    
    while start < end: 
        # Move start index to the right while arr[start] <= pivot
        while start < ub and arr[start] <= p:  # Add bounds check for start
            start += 1
        # Move end index to the left while arr[end] > pivot
        while end > lb and arr[end] > p:  # Add bounds check for end
            end -= 1
        # Swap if the pointers have not crossed
        if start < end: 
            arr[start], arr[end] = arr[end], arr[start]
    
    # Place the pivot in its correct position
    arr[lb], arr[end] = arr[end], arr[lb]
    return end


def quickSort(arr, lb, ub):
    if lb < ub:
        loc = partition(arr, lb, ub)
        quickSort(arr, lb, loc - 1)  # Sort elements before partition
        quickSort(arr, loc + 1, ub)   # Sort elements after partition
        return arr


# Example usage
if __name__ == "__main__":
    arr = [64, 34, 25, 12, 22, 11, 90]
    print("Unsorted array:", arr)
    quickSort(arr, 0, len(arr) - 1)
    print("Sorted array:", arr)
```
